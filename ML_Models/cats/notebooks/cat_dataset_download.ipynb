## 🔍 Step 5: Compare Dataset Structures

Let's explore and compare the folder structures and contents of both downloaded datasets to understand their differences.

```python
def explore_directory_structure(path, max_depth=3, current_depth=0):
    """Recursively explore directory structure"""
    items = []
    if current_depth >= max_depth:
        return items
    
    try:
        for item in sorted(os.listdir(path)):
            item_path = os.path.join(path, item)
            indent = "  " * current_depth
            if os.path.isdir(item_path):
                items.append(f"{indent}📁 {item}/")
                items.extend(explore_directory_structure(item_path, max_depth, current_depth + 1))
            else:
                items.append(f"{indent}📄 {item}")
    except PermissionError:
        items.append(f"{indent}❌ Permission denied")
    
    return items

print("🔍 OXFORD-IIIT PET DATASET STRUCTURE:")
print("=" * 50)
oxford_structure = explore_directory_structure(oxford_destination)
for item in oxford_structure[:20]:  # Show first 20 items
    print(item)
if len(oxford_structure) > 20:
    print(f"... and {len(oxford_structure) - 20} more items")

print("\n🐱 CAT BREEDS DATASET STRUCTURE:")
print("=" * 50)
cat_breeds_structure = explore_directory_structure(cat_breeds_destination)
for item in cat_breeds_structure[:20]:  # Show first 20 items
    print(item)
if len(cat_breeds_structure) > 20:
    print(f"... and {len(cat_breeds_structure) - 20} more items")
```

## 📊 Step 6: Explore Dataset Contents

Let's examine the contents of both datasets in detail, including image counts, file formats, and breed information.

```python
def analyze_dataset(dataset_path, dataset_name):
    """Analyze dataset contents and structure"""
    print(f"\n📊 ANALYZING {dataset_name.upper()}")
    print("=" * 60)
    
    total_files = 0
    image_files = 0
    folders = []
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp'}
    
    for root, dirs, files in os.walk(dataset_path):
        folders.extend(dirs)
        for file in files:
            total_files += 1
            if any(file.lower().endswith(ext) for ext in image_extensions):
                image_files += 1
    
    # Count unique folders (potential breeds)
    unique_folders = set(folders)
    
    print(f"📁 Total folders found: {len(unique_folders)}")
    print(f"📄 Total files: {total_files}")
    print(f"🖼️ Image files: {image_files}")
    print(f"📊 Other files: {total_files - image_files}")
    
    # Show some folder names (potential breed names)
    if unique_folders:
        print(f"\n🏷️ Sample folder names (potential breeds):")
        for i, folder in enumerate(sorted(unique_folders)[:10]):
            print(f"   {i+1}. {folder}")
        if len(unique_folders) > 10:
            print(f"   ... and {len(unique_folders) - 10} more")
    
    return {
        'total_files': total_files,
        'image_files': image_files,
        'folders': len(unique_folders),
        'folder_names': sorted(unique_folders)
    }

# Analyze both datasets
oxford_stats = analyze_dataset(oxford_destination, "Oxford-IIIT Pet Dataset")
cat_breeds_stats = analyze_dataset(cat_breeds_destination, "Cat Breeds Dataset")
```

## ✅ Step 7: Verify Downloaded Files and Choose Best Dataset

Let's verify the downloads were successful and determine which dataset is better for our cat breed classification model.

```python
# Compare datasets and make recommendation
print("🎯 DATASET COMPARISON SUMMARY")
print("=" * 60)

datasets = [
    ("Oxford-IIIT Pet Dataset", oxford_stats),
    ("Cat Breeds Dataset", cat_breeds_stats)
]

for name, stats in datasets:
    print(f"\n📊 {name}:")
    print(f"   🖼️ Images: {stats['image_files']:,}")
    print(f"   📁 Categories: {stats['folders']}")
    print(f"   📄 Total files: {stats['total_files']:,}")
    print(f"   📊 Images per category: {stats['image_files'] // max(stats['folders'], 1)}")

print("\n🏆 RECOMMENDATION:")
print("=" * 30)

# Make recommendation based on the stats
if cat_breeds_stats['image_files'] > oxford_stats['image_files']:
    recommended = "Cat Breeds Dataset"
    reason = "More images available for training"
elif cat_breeds_stats['folders'] > oxford_stats['folders']:
    recommended = "Cat Breeds Dataset"
    reason = "More breed categories available"
else:
    recommended = "Oxford-IIIT Pet Dataset"
    reason = "Well-established research dataset"

print(f"✅ Recommended dataset: {recommended}")
print(f"💡 Reason: {reason}")

print(f"\n📍 Dataset locations:")
print(f"   Oxford-IIIT: {oxford_destination}")
print(f"   Cat Breeds: {cat_breeds_destination}")

print(f"\n🎉 Setup complete! You can now proceed with data preprocessing and model training.")
```
