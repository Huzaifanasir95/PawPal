# Download Stanford Dogs Dataset

This notebook downloads the Stanford Dogs Dataset from Stanford Vision Lab.

**Dataset Information:**
- **Source:** http://vision.stanford.edu/aditya86/ImageNetDogs/
- **Number of categories:** 120 dog breeds
- **Number of images:** 20,580
- **Annotations:** Class labels and Bounding boxes

**Components to download:**
1. Images (757MB)
2. Annotations (21MB)
3. Train/Test split lists (0.5MB)

````python
import os
import urllib.request
import tarfile
from pathlib import Path
from tqdm import tqdm

# Define data directory
DATA_DIR = Path('../data/stanford_dogs')
DATA_DIR.mkdir(parents=True, exist_ok=True)

print(f"Data will be saved to: {DATA_DIR.absolute()}")
````

## Download Helper Function

This function downloads files with a progress bar.

````python
class DownloadProgressBar(tqdm):
    def update_to(self, b=1, bsize=1, tsize=None):
        if tsize is not None:
            self.total = tsize
        self.update(b * bsize - self.n)

def download_url(url, output_path):
    """
    Download a file from a URL with progress bar.
    
    Args:
        url (str): URL to download from
        output_path (Path): Path to save the downloaded file
    """
    print(f"\nDownloading from: {url}")
    print(f"Saving to: {output_path}")
    
    with DownloadProgressBar(unit='B', unit_scale=True, miniters=1, desc=output_path.name) as t:
        urllib.request.urlretrieve(url, filename=output_path, reporthook=t.update_to)
    
    print(f"✓ Downloaded: {output_path.name}")
````

## 1. Download Images (757MB)

This contains all 20,580 dog images organized by breed.

````python
# Download images
images_url = "http://vision.stanford.edu/aditya86/ImageNetDogs/images.tar"
images_tar_path = DATA_DIR / "images.tar"

if not images_tar_path.exists():
    download_url(images_url, images_tar_path)
else:
    print(f"✓ Images archive already exists: {images_tar_path}")
````

## 2. Download Annotations (21MB)

This contains bounding box annotations for all images.

````python
# Download annotations
annotations_url = "http://vision.stanford.edu/aditya86/ImageNetDogs/annotation.tar"
annotations_tar_path = DATA_DIR / "annotation.tar"

if not annotations_tar_path.exists():
    download_url(annotations_url, annotations_tar_path)
else:
    print(f"✓ Annotations archive already exists: {annotations_tar_path}")
````

## 3. Download Train/Test Splits (0.5MB)

This contains the official train and test split lists.

````python
# Download train/test splits
lists_url = "http://vision.stanford.edu/aditya86/ImageNetDogs/lists.tar"
lists_tar_path = DATA_DIR / "lists.tar"

if not lists_tar_path.exists():
    download_url(lists_url, lists_tar_path)
else:
    print(f"✓ Lists archive already exists: {lists_tar_path}")
````

## Extract Downloaded Files

Now let's extract all the tar archives.

````python
def extract_tar(tar_path, extract_to=None):
    """
    Extract a tar archive.
    
    Args:
        tar_path (Path): Path to the tar file
        extract_to (Path): Directory to extract to (default: same as tar file location)
    """
    if extract_to is None:
        extract_to = tar_path.parent
    
    print(f"\nExtracting: {tar_path.name}...")
    
    with tarfile.open(tar_path, 'r') as tar:
        tar.extractall(path=extract_to)
    
    print(f"✓ Extracted to: {extract_to}")

# Extract all archives
print("="*50)
print("Starting extraction process...")
print("="*50)

# Extract images
if images_tar_path.exists():
    images_dir = DATA_DIR / "Images"
    if not images_dir.exists():
        extract_tar(images_tar_path, DATA_DIR)
    else:
        print(f"✓ Images already extracted: {images_dir}")

# Extract annotations
if annotations_tar_path.exists():
    annotations_dir = DATA_DIR / "Annotation"
    if not annotations_dir.exists():
        extract_tar(annotations_tar_path, DATA_DIR)
    else:
        print(f"✓ Annotations already extracted: {annotations_dir}")

# Extract lists
if lists_tar_path.exists():
    lists_files = list(DATA_DIR.glob("*list.mat"))
    if len(lists_files) == 0:
        extract_tar(lists_tar_path, DATA_DIR)
    else:
        print(f"✓ Lists already extracted")

print("\n" + "="*50)
print("Extraction complete!")
print("="*50)
````

## Verify Dataset Structure

Let's verify that everything was downloaded and extracted correctly.

````python
# Check dataset structure
print("\n" + "="*50)
print("Dataset Structure:")
print("="*50)

images_dir = DATA_DIR / "Images"
annotations_dir = DATA_DIR / "Annotation"

if images_dir.exists():
    breed_dirs = [d for d in images_dir.iterdir() if d.is_dir()]
    total_images = sum(len(list(d.glob("*.jpg"))) for d in breed_dirs)
    print(f"✓ Images directory: {images_dir}")
    print(f"  - Number of breeds: {len(breed_dirs)}")
    print(f"  - Total images: {total_images}")
else:
    print("✗ Images directory not found")

if annotations_dir.exists():
    annotation_dirs = [d for d in annotations_dir.iterdir() if d.is_dir()]
    total_annotations = sum(len(list(d.glob("*"))) for d in annotation_dirs)
    print(f"\n✓ Annotations directory: {annotations_dir}")
    print(f"  - Number of breeds: {len(annotation_dirs)}")
    print(f"  - Total annotations: {total_annotations}")
else:
    print("\n✗ Annotations directory not found")

# Check for list files
list_files = list(DATA_DIR.glob("*list.mat"))
if list_files:
    print(f"\n✓ Found {len(list_files)} list files:")
    for f in list_files:
        print(f"  - {f.name}")
else:
    print("\n✗ List files not found")

print("\n" + "="*50)
print("Dataset download and extraction completed successfully!")
print("="*50)
````
